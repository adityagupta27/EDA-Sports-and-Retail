## Correlation of Dataset

As the name itself says that 'Co' means “together” and then Relation . Correlation is the expression of relation between two variables means the rate at which they both will change with each other as these variables are dependent to each other.

```python
retail_corr = retail.corr()
retail_corr
```

```python
# heat-map of correlation
axes = plt.subplots(1,1, figsize = (10,6))
sb.heatmap(retail_corr,annot = True,cmap = 'Oranges')
plt.show()
```

### Covariance of Dataset

Covariance is a measure of how much two random variables vary together. It’s similar to variance, but where variance tells you how a single variable varies, co variance tells you how two variables vary together. Cov(X,Y) = Σ E((X-μ)E(Y-ν)) / n-1 where: X is a random variable E(X) = μ is the expected value (the mean) of the random variable X and E(Y) = ν is the expected value (the mean) of the random variable Y n = the number of items in the data set

```python
retail_cov = retail.cov()
retail_cov
```

```python
# heat-map of Covariance 
axes = plt.subplots(1,1, figsize = (10,6))
sb.heatmap(retail_cov,annot = True)
plt.show()
```

## Bar chart or Bar graph
It is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

```python
plt.figure(figsize = (10,8))
plt.bar(x ='Sub-Category',height = 'Category', data = retail )
plt.title('Sub Category in diff. Catogory', size = 30, pad=10 )
plt.xlabel('Sub Category')
plt.ylabel('Category')
plt.xticks(rotation = 45)
plt.show()
```

# Scatter-Plot

scatter plot is a diagram where each value in the data set is represented by a dot.
The Matplotlib module has a method for drawing scatter plots, it needs two arrays of the same length, 
one for the values of the x-axis, and one for the values of the y-axis

```python
retail.plot.scatter(x = 'Discount', y='Profit', color ='green', marker = 'o', alpha = 0.4)
```

 Business Insights : answer us in understanding the relationship between discount and profit ,that from the graph we get to kow that when there is less discount we have more profit but when we gives high discount then we got less profit.

# hist()
function is defined as a quick way to understand the distribution of certain numerical variables from the dataset. It divides the values within a numerical variable into "bins". It counts the number of examinations that fall into each of the bin. These bins are responsible for a rapid and intuitive sense of the distribution of the values within a variable by visualizing bins

```python
retail.hist(bins = 10, figsize =(20,15), facecolor ='green', alpha = 0.4)
```
